```python
def create_driver(download_directory):
    # 创建浏览器选项
    options = Options()
    options.page_load_strategy = 'normal'  # 设置页面加载策略
    # 配置下载路径
    options.add_experimental_option("prefs", {
        "download.default_directory": download_directory,
        "download.prompt_for_download": False,
        "download.directory_upgrade": True
    })
    # 返回一个新的浏览器实例
    return webdriver.Edge(service=edge_service, options=options)

def download_data(year, month):
    # 创建一个新的浏览器实例并指定下载目录
    driver = create_driver("D:\\data\\flight_data")
    driver.get("https://www.transtats.bts.gov/DL_SelectFields.aspx?gnoyr_VQ=FGK&QO_fu146_anzr=b0-gvzr")
    driver.set_page_load_timeout(1800)  # 设置页面加载超时时间为 1800 秒
    driver.set_script_timeout(1800)     # 设置脚本执行超时时间为 1800 秒
    
    # 选择年份
    year_select = Select(driver.find_element(By.NAME, "cboYear"))
    year_select.select_by_value(str(year))
    
    # 选择月份
    period_select = Select(driver.find_element(By.NAME, "cboPeriod"))
    period_select.select_by_value(str(month))
    
    # 选择地区为 "All"
    geography_select = Select(driver.find_element(By.NAME, "cboGeography"))
    geography_select.select_by_value("All")
    
    # 选择全部特征
    for feature_ID in ID_list:
        #time.sleep(0.1)
        try:
            checkbox = driver.find_element(By.ID, feature_ID)
            if not checkbox.is_selected():
                checkbox.click()
        except Exception as e:
            print(f"勾选{feature_ID}出错:", e)
            
    # 从页面底部逐步滚动到顶部
    for i in range(0, driver.execute_script("return document.body.scrollHeight;"), 100):
        driver.execute_script(f"window.scrollTo(0, document.body.scrollHeight - {i});")
        time.sleep(0.1)  # 每次滚动后稍作等待


    # 等待并点击下载按钮
    try:
        download_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.ID, "btnDownload"))
        )
        download_button.click()
    except Exception as e:
        print("下载报错:", e)
        
    #等待页面创建下载任务
    print('wait for 300 s before start another download job')
    for i in range(30):
        time.sleep(10)
        print(f"{i+1}0/300 s")
    # 完成后关闭浏览器
    driver.quit()

# 主函数
def main():
    # 定义年份和月份范围
    years = [2021,2022,2023,2024]  
    months = [1,11,12]
    #months = range(1, 13)  # 1 到 12 代表各个月份
    for year in years:
        for month in months:
            download_data(year, month)
            print(f"Creat download job {year} - {month:02d}")
main()
```
